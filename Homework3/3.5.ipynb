```python
import random as rd
import sys
import time

sys.setrecursionlimit(100000)

def insertion_sort(A):
    n=len(A)
    for i in range(1,n):
        j=i-1
        while j>0 and A[j+1]<A[j]:
            A[j],A[j+1]=A[j+1],A[j]
            j-=1

def selection_sort(A):
    n=len(A)
    for i in range(0,n):
        for j in range(i+1,n):
            if A[j]<A[i]:
                A[j],A[i]=A[i],A[j]
                
def partition(A,p,r):
    x=A[r]
    i=p-1
    for j in range(p,r):
        if A[j]<=x:
            i+=1
            A[i],A[j]=A[j],A[i]
    A[i+1],A[r]=A[r],A[i+1]
    return i+1

def quick_sort(A,p,r):
    if p<r:
        q=partition(A,p,r)
        quick_sort(A,p,q-1)
        quick_sort(A,q+1,r)

List100=[]
List1000=[]
List10000=[]

for i in range(0,100):
    List100.append(rd.randint(1,10))
for i in range(0,1000):
    List1000.append(rd.randint(1,10))
for i in range(1,10000):
    List10000.append(rd.randint(1,10))

start1=time.time()
insertion_sort(List100)
end1=time.time()
time1=end1-start1
print("Insertion sort took", time1, "s for a list of 100 numbers.")
start2=time.time()
insertion_sort(List1000)
end2=time.time()
time2=end2-start2
print("Insertion sort took", time2, "s for a list of 1000 numbers.")
start3=time.time()
insertion_sort(List10000)
end3=time.time()
time3=end3-start3
print("Insertion sort took", time3, "s for a list of 10000 numbers.")

start4=time.time()
selection_sort(List100)
end4=time.time()
time4=end4-start4
print("Selction sort took", time4, "s for a list of 100 numbers.")
start5=time.time()
selection_sort(List1000)
end5=time.time()
time5=end5-start5
print("Selecttion sort took", time5, "s for a list of 1000 numbers.")
start6=time.time()
selection_sort(List10000)
end6=time.time()
time6=end6-start6
print("Selecttion sort took", time6, "s for a list of 10000 numbers.")

start7=time.time()
quick_sort(List100,0,len(List100)-1)
end7=time.time()
time7=end7-start7
print("Quick sort took", time7, "s for a list of 100 numbers.")
start8=time.time()
quick_sort(List1000,0,len(List1000)-1)
end8=time.time()
time8=end8-start8
print("Quick sort took", time8, "s for a list of 1000 numbers.")
start9=time.time()
quick_sort(List10000,0,len(List10000)-1)
end9=time.time()
time9=end9-start9
print("Quick sort took", time9, "s for a list of 10000 numbers.")
```
